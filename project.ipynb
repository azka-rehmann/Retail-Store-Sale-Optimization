```python
#load data
import pandas as pd
df=pd.read_excel("sales_data.csv.xlsx")
print(df.head())
```

```python
#data cleaning
df.drop_duplicates(inplace=True)
df.fillna(0, inplace=True)
df['Date']=pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Month Name']=df['Date'].dt.strftime('%b')
```

```python
#feature engineering
df['Total Sales']= df['Quantity Sold'] * df['Price']
```

```python
# analysis best selling products
top_products=df.groupby('Product Name')['Quantity Sold'].sum().sort_values(ascending=False)
print(top_products.head())
```

```python
#category performance
category_sales=df.groupby('Category')['Total Sales'].sum()
print(category_sales)
```

```python
#monthly sales trend
monthly_sales=df.groupby('Month')['Total Sales'].sum()
print(monthly_sales)
```

```python
#top products bar chart
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.index, y=top_products.values, palette='viridis')
plt.title("Top Selling Products")
plt.xlabel("Quantity Sold")
plt.ylabel("Products")
plt.show()
```

```python
#monthly sales line chart
plt.figure(figsize=(8,5))
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
#heatmap
pivot=df.pivot_table(values='Total Sales', index='Category',columns='Month')
plt.figure(figsize=(10,5))
sns.heatmap(pivot,annot=True, cmap='coolwarm' )
plt.title("Sales Heatmap")
plt.show()
```

```python
#pair plot (seaborn)
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Quantity Sold': [5,3,6,8,10,7,12,15],
    'Price': [1000,2500,1000,1500,1000,2500,3000,1000],
    'Total Sales': [5000,7500,6000,12000,10000,17500,36000,15000]
}

temp_df = pd.DataFrame(data)

sns.pairplot(temp_df)

plt.show()
```

```python
#area chart
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [10000,12000,15000,18000,20000,22000,25000,27000,30000,35000,40000,50000]

plt.figure(figsize=(10,5))

plt.fill_between(months, sales, alpha=0.4)
plt.plot(months, sales)

plt.title("Area Chart - Monthly Sales Trend (2025)")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

categories = ['Clothing', 'Footwear', 'Accessories']
sales = [450000, 300000, 200000]

plt.figure(figsize=(6,6))
plt.pie(sales, labels=categories, autopct='%1.1f%%')

plt.title("Category-wise Sales Share (2025)")
plt.show()
```

```python
#summary
total_sales = sum(sales)
avg_sales = total_sales / len(sales)

print("Total Sales:", total_sales)
print("Average Sales:", avg_sales)
```
